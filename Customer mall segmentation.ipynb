# IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import seaborn as sns
```

# IMPORT THE DATA SET

```python
data = pd.read_csv('Mall_Customers.csv')
data.head()
```

# Use 2 columns as the basic features for our clustering

```python
data_clus = data.iloc[:, 3:5]
data_clus.head()
```

# Plot the initial data of all 200 customers

```python
plt.scatter(data_clus.iloc[:,0], data_clus.iloc[:,1],c='black')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
clus_array = np.array(data_clus)
print (clus_array)
```

# Use the MinMaxScaler

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(clus_array)
data_scaled
```

#Set the range to predict the value of k in the range of 1 to 10.

```python
distortions = []
K = range(1,10)
for k in K:
  kmeanModel = KMeans(n_clusters = k)
  kmeanModel.fit(data_scaled)
  distortions.append(kmeanModel.inertia_)
```

# See the distortion in the visualization

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
```

# Create a model using our scaled data

```python
kmeanModel = KMeans(n_clusters=5)
kmeanModel.fit(data_scaled)
```

```python
print(kmeanModel.cluster_centers_)
```

# Visualize the position of the centroids

```python
plt.scatter(data_scaled[:,0], data_scaled[:,1], s=50, c='black', marker='o', alpha=1)
centers = kmeanModel.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='red', s=50, alpha=1, marker='o')
plt.title('K-Means Centroid')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

# Predict our scaled data using the K-Means Model

```python
y_kmeans = kmeanModel.predict(data_scaled)

plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeanModel.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

# Create a new dataframe of the cluster results.

```python
result = pd.DataFrame(data = y_kmeans)
result.columns=['cluster']
result.head()
```

# Combine the results with the initial dataset

```python
output = pd.concat([data, result], axis=1)
output.head()
```

# re-plot the grouping results by adding legend information using the seaborn library

```python
sns.scatterplot(x=output.iloc[:,3], y=output.iloc[:,4], hue=output.iloc[:,5], data=output, palette='viridis', s=40, legend="full");
plt.legend(loc='center right', fontsize='10');
```

```python
output.to_csv('out.csv', index=False)
```
